# NumPy (Numerical Python)

**NumPy** is a fundamental library for numerical computing in Python, created in **2005**.

Key features:
- **ndarray**: N-dimensional array object — fast, memory-efficient container for large datasets
- **Vectorization**: Operations on entire arrays without writing loops (much faster than Python lists)
- **Broadcasting**: Rules for applying operations on arrays of different shapes

````python
import numpy as np
````

````python
lst = [12, 15, 7, 20, 18]   # Create a Python list
a = np.array(lst) # Convert the list to a NumPy array
print(a)
````

````python
# type() shows the data type of the variable
print(type(a))
````

## Fancy Indexing

Selecting elements using a **list or array of indices** (not slices).

- Select specific **rows**: `b[[0, 2]]` → rows 0 and 2
- Select specific **elements**: `b[[r1, r2], [c1, c2]]` → elements at (r1,c1) and (r2,c2)

This is different from basic slicing — fancy indexing always returns a **copy**, not a view.

````python
b = np.array([[12, 15, 7, 20, 18],
              [14, 16, 15, 13, 3],
              [4, 6, 5, 99, 23],
              [7, 8, 12, 11, 3]])

# Select rows 0 and 2 using a list of indices
print(b[[0, 2]])
````

````python
# Select specific elements: b[0,1], b[1,3], b[2,4]
print(b[[0, 1, 2], [1, 3, 4]])
````

### Understanding `b[:2, :]`

This is **2D slicing** with two parts separated by a comma: `array[row_slice, col_slice]`

| Part | Meaning |
|---|---|
| `:2` | **Rows**: from start up to (not including) index 2 → rows **0** and **1** |
| `:` | **Columns**: all columns (no filtering) |

```
b = [[12, 15,  7, 20, 18],   ← row 0  ✅ selected
     [14, 16, 15, 13,  3],   ← row 1  ✅ selected
     [ 4,  6,  5, 99, 23],   ← row 2  ✗
     [ 7,  8, 12, 11,  3]]   ← row 3  ✗

b[:2, :] → [[12, 15,  7, 20, 18],
            [14, 16, 15, 13,  3]]
```

````python
b[:2, :] = 0 # Set the first two rows to zero
print(b)
````

````python
b[[0,2]] = 100 # Set rows 0 and 2 to 100
print(b)
````

## Element-wise Operations

NumPy performs arithmetic **element by element** on arrays of the same shape:
- `x + y` → adds corresponding elements
- `x * y` → multiplies corresponding elements
- `x / y` → divides corresponding elements

This is called **vectorization** and is much faster than looping through elements manually.

````python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
````

````python
# Element-wise addition: [1+4, 2+5, 3+6]
z = x + y
print("x + y =", z)
````

````python
# Element-wise multiplication: [1*4, 2*5, 3*6]
print("x * y =", x * y)
````

````python
# Element-wise division: [1/4, 2/5, 3/6]
print("x / y =", x / y)
````

````python
# Broadcasting: scalar 10 is applied to every element
print("x * 10 =", x * 10)
````

````python
# sqrt: square root of each element
print("sqrt(x) =", np.sqrt(x))
````

## Padding Arrays (`np.pad`)

`np.pad(array, (before, after))` adds zeros to the beginning or end of an array.

This is useful when you need to **add arrays of different lengths** — pad the shorter one with zeros first so both arrays have the same size.

````python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([11, 15, 16])
````

````python
# Pad shorter array with zeros, then add
result = np.pad(x, (0, max(0, len(y)-len(x)))) + np.pad(y, (0, max(0, len(x)-len(y))))
print(result)
````

## Broadcasting

NumPy's mechanism for performing operations on arrays with **different shapes**.

The smaller array is automatically "broadcast" (stretched) to match the larger array.

**Rules:**
1. Arrays with fewer dimensions get padded with 1s on the left
2. Dimensions of size 1 are stretched to match the other array
3. If dimensions don't match and neither is 1, an error is raised

````python
a = np.array([[1, 2, 3], [4, 5, 6]])  # shape (2, 3)
b = np.array([10, 20, 30])             # shape (3,) → broadcast to (2, 3)

# b is automatically broadcast to match a's shape
print(a + b)
````

````python
# Column vector (3,1) + Row vector (1,2) → result (3,2)
col = np.array([[1], [2], [3]])  # shape (3, 1)
row = np.array([10, 20])         # shape (2,)

print(col + row)
````

````python
x = np.array([1, -2, 3, -4, 5])

# np.where: replace negatives with 0
print(np.where(x > 0, x, 0))
````

````python
# ufunc with where parameter: only compute sqrt where x > 0
result = np.zeros_like(x, dtype=float)
np.sqrt(x, where=(x > 0), out=result)
print(result)
````

## NumPy vs Python List

Why use NumPy arrays instead of Python lists?

1. **Speed**: NumPy stores data in **contiguous memory** → much faster (C-level optimized)
2. **Vectorized operations**: Direct math on arrays without Python loops
3. **Less memory**: Fixed `dtype` → lower memory usage, ideal for **ML/DL**

````python
import time

# Speed comparison: Python list vs NumPy array
size = 1_000_000
lst = list(range(size))
arr = np.arange(size)

start = time.time()
result_list = [x * 2 for x in lst]
list_time = time.time() - start

start = time.time()
result_arr = arr * 2
numpy_time = time.time() - start

print(f"Python list: {list_time:.4f}s")
print(f"NumPy array: {numpy_time:.4f}s")
print(f"NumPy is ~{list_time / numpy_time:.0f}x faster!")
````

````python
# Simpler syntax with NumPy

# Python list: need list comprehension
lst = [1, 2, 3, 4, 5]
result_list = [x ** 2 + 2 * x + 1 for x in lst]
print("List:", result_list)

# NumPy: direct mathematical expression
arr = np.array([1, 2, 3, 4, 5])
result_numpy = arr ** 2 + 2 * arr + 1
print("NumPy:", result_numpy)
````

## Linear Algebra (`np.linalg`)

### Vector Norm
Norm measures the **magnitude** (size) of a vector — how far it is from the origin.

- **L1 norm**: sum of absolute values → `|x₁| + |x₂| + ...`
- **L2 norm**: Euclidean distance → `√(x₁² + x₂² + ...)`

````python
v = np.array([3, -4])

print("L1 norm:", np.linalg.norm(v, ord=1))  # |3| + |-4| = 7
print("L2 norm:", np.linalg.norm(v, ord=2))  # √(9 + 16) = 5.0
````

### Determinant & Inverse Matrix

- **Determinant**: A scalar value computed from a square matrix. If `det = 0`, the matrix has no inverse.
- **Inverse**: `A⁻¹` such that `A @ A⁻¹ = Identity matrix`

````python
A = np.array([[1, 2], [3, 4]])

# Determinant: 1*4 - 2*3 = -2
print("det(A):", np.linalg.det(A))
````

````python
# Inverse matrix
A_inv = np.linalg.inv(A)
print("A inverse:\n", A_inv)

# Verify: A @ A_inv ≈ Identity
print("\nA @ A_inv:\n", np.round(A @ A_inv))
````

### Eigenvalues & Eigenvectors

When a matrix is applied to a special vector (**eigenvector**), it only **scales** the vector — doesn't change its direction. The scaling factor is the **eigenvalue**.

`A @ v = λ × v` where `v` = eigenvector, `λ` = eigenvalue

````python
A = np.array([[2, 1], [1, 2]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
````

````python
# Verify: A @ eigenvector = eigenvalue * eigenvector
v = eigenvectors[:, 0]  # type: ignore # first eigenvector
lam = eigenvalues[0]    # first eigenvalue

print("A @ v   =", A @ v)
print("λ × v   =", lam * v)
print("Equal?", np.allclose(A @ v, lam * v))
````
