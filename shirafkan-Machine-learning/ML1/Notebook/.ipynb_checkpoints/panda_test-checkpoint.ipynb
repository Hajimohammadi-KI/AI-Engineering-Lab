Sure — here is a **clean English rewrite with clear examples**, suitable for **exam notes** 👇

---

## Important Comparison for the Exam

| Question goal                                                 | Correct syntax      | Example                          |
| ------------------------------------------------------------- | ------------------- | -------------------------------- |
| Select data using **labels** (row/column names)               | `df2.loc[...]`       | `df2.loc[0, 'MedHouseVal']`       |
| Select data using **integer positions**                       | `df2.iloc[...]`      | `df2.iloc[0, 7]`                  |
| Select the **first 5 rows and first 3 columns** (by position) | `df2.iloc[0:5, 0:3]` | Returns rows 0–4 and columns 0–2 |

---

## Key Rules to Remember (Exam-friendly)

* **`.loc` → label-based indexing**
  Uses **row/column names**

* **`.iloc` → integer position-based indexing**
  Uses **row/column numbers (0-based)**

---

### Example DataFrame

```python
df2.columns
# ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms',
#  'Population', 'AveOccup', 'Latitude', 'Longitude', 'MedHouseVal']
```

### Examples

```python
df2.loc[0, 'MedHouseVal']     # label-based (column name)
df2.iloc[0, 8]               # position-based (9th column)
df2.iloc[0:5, 0:3]           # first 5 rows, first 3 columns
```

---

### One-line exam takeaway ✅

> **Use `.loc` for labels, use `.iloc` for integer positions.**
