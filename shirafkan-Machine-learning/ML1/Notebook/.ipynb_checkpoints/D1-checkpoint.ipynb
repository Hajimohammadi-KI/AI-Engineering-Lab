````python
from sklearn.datasets import fetch_california_housing
from sklearn.utils import Bunch

data: Bunch = fetch_california_housing()
data.target  # Kein Fehler mehr
````

````python
# California Housing dataset does not have target_names attribute
# This dataset is for regression (continuous values), not classification
# data.target_names  # This would cause an error
print("Target: Median house value (in $100,000s)")
````

````python
# Display feature names (column names for the dataset)
# Features: MedInc, HouseAge, AveRooms, AveBedrms, Population, AveOccup, Latitude, Longitude
data.feature_names  # type: ignore
````

````python
# Create a DataFrame from the feature data with proper column names
# This makes the data easier to work with and visualize
df = pd.DataFrame(data.data, columns=data.feature_names) # type: ignore
df
````

````python
# Check for missing values in each column
# Returns count of NaN values per column (should be 0 for this dataset)
df.isnull().sum()
````

````python
# Add the target variable (median house value) as a new column to the DataFrame
df['Target'] = data.target
df
````

````python
# Select and display only the 'HouseAge' column
# Double brackets [[]] return a DataFrame, single brackets [] return a Series
df[['HouseAge']]
````

````python
# Select multiple columns at once
# Returns a DataFrame with only HouseAge, AveBedrms, and Target columns
df[['HouseAge', 'AveBedrms', 'Target']]
````

````python
# Display statistical summary of the DataFrame
# Shows count, mean, std, min, 25%, 50%, 75%, max for each column
df.describe()
````

````python
# Sort DataFrame by 'HouseAge' column in ascending order (default)
# Oldest houses will be at the bottom
df.sort_values(by='HouseAge')
````

````python
# Sort DataFrame by 'HouseAge' column in descending order
# Oldest houses will be at the top
df.sort_values(by='HouseAge', ascending=False)
````

````python
# Display the first 5 rows of the DataFrame
# Useful for quick data inspection
df.head()
````

````python
# Display the last 5 rows of the DataFrame
# Useful for checking the end of the data
df.tail()
````

````python
# Display the entire DataFrame
df
````

````python
# Filter rows where Target (house value) is greater than $450,000
# Boolean indexing to select expensive houses
df[df['Target'] > 4.5]
````

````python
# Filter rows where Target is between $300,000 and $400,000
# Using multiple conditions with & (AND) operator
# Note: Each condition must be in parentheses
df[(df['Target'] > 3) & (df['Target'] < 4)]
````

````python
# Display the DataFrame again (unchanged)
df
````

````python
# Import NumPy again (for creating NaN values)
import numpy as np
````

````python
# Create a copy of the DataFrame to practice handling missing values
# Using copy() to avoid modifying the original DataFrame
df2 = df.copy()
df2
````

````python
# Intentionally create missing values (NaN) in 'HouseAge' column
# Using .loc[row, column] to set specific cells to NaN
df2.loc[0, 'HouseAge'] = np.nan      # First row
df2.loc[2, 'HouseAge'] = np.nan      # Third row
df2.loc[20638, 'HouseAge'] = np.nan  # Second to last row
df2.loc[20639, 'HouseAge'] = np.nan  # Last row

df2
````

````python
# Add another missing value in 'Population' column
df2.loc[20639, 'Population'] = np.nan
df2
````

````python
# Display df2 with the missing values we created
df2
````

````python
# Check which cells are null (missing)
# Returns a DataFrame of True/False values
# True = missing value, False = valid value
df2.isnull()
````

````python
# Count the number of missing values in each column
# HouseAge should have 4, Population should have 1
df2.isnull().sum()  # Alternative: df2.isna().sum()
````

````python
# Calculate the mean of 'HouseAge' column (ignoring NaN values)
# This value will be used to fill missing values
m = df2['HouseAge'].mean()
m
````

````python
# Better approach: fill each column with its own mean
# This fills HouseAge NaN with HouseAge mean, Population NaN with Population mean
df3 = df2.fillna(df2.mean())
df3
````

````python
# Alternative approach: fill specific column with its own mean
# This only fills Population NaN values, leaving other columns unchanged
df3["Population"] = df2["Population"].fillna(df2["Population"].mean())
df3
````

````python
# Sort df3 by 'HouseAge' column in descending order
# Shows oldest houses first
df3.sort_values(by='HouseAge', ascending=False)
````
