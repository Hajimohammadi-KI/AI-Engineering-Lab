## Why the error occurs (simply explained)

`load_iris()` can have two different return types:

- **Bunch object** (default): contains metadata such as `feature_names`

- **Tuple `(X, y)`** when `return_X_y=True`: contains **only** data and labels, **no** `feature_names`

Therefore, we first check the type and only read `feature_names` if a Bunch is present.

```python
`iris_bunch = data_iris[0] if isinstance(data_iris, tuple) else data_iris
`print(iris_bunch.feature_names)`

```

Additionally, `iris_bunch.feature_names` is clearer than `iris_bunch["feature_names"]` and is better recognized by Pylance.

````python
# If data_iris is a tuple (X, y), it does not contain feature_names.
# In that case, load the Bunch object separately so feature_names is available.
if isinstance(data_iris, tuple):
	iris_bunch = load_iris(return_X_y=False)
else:
	iris_bunch = data_iris

print(iris_bunch.feature_names)
````
