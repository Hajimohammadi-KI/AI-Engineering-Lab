```python
# Create a copy of the DataFrame
df2 = df.copy()
# Display the copied DataFrame
df2
```

```python
# Set some values in 'HouseAge' column to NaN (missing)
df2.loc[0,'HouseAge'] = np.nan
df2.loc[2,'HouseAge'] = np.nan
df2.loc[20638,'HouseAge'] = np.nan
df2.loc[20639,'HouseAge'] = np.nan

# Display the DataFrame with NaN values
df2
```

```python
# Check for missing values in the DataFrame
df2.isnull()
```

```python
# Count the number of missing values in each column
df2.isnull().sum()   # or use df2.isna().sum()
```

```python
# Calculate the mean of the 'HouseAge' column
m = df2['HouseAge'].mean()
# Show the calculated mean
m
```

```python
# Fill missing values in the DataFrame with the mean of 'HouseAge'
df3 = df2.fillna(m)
# Display the DataFrame after filling missing values
df3
```

```python
# Count the number of missing values in each column after filling
df3.isnull().sum()
```
