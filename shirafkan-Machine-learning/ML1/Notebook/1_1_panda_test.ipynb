## Summary: .loc vs .iloc

| Goal | Syntax | Example |
|------|--------|---------|
| Select by **labels** (column/row names) | `df.loc[...]` | `df.loc[0, 'MedHouseVal']` |
| Select by **integer positions** | `df.iloc[...]` | `df.iloc[0, 7]` |
| Select first 5 rows, first 3 columns | `df.iloc[0:5, 0:3]` | Returns rows 0–4 and columns 0–2 |

---

### Key Rules
- **`.loc` uses labels** (row/column names) - END IS INCLUSIVE
- **`.iloc` uses integer positions** (0-based) - END IS EXCLUSIVE

---

### Example DataFrame
```
df2.columns
# Output: ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 
#          'Population', 'AveOccup', 'Latitude', 'Longitude', 'MedHouseVal']
```

### Examples
```python
df2.loc[0, 'MedHouseVal']     # Get value at row 0, column 'MedHouseVal' (by label)
df2.iloc[0, 8]               # Get value at row index 0, column index 8 (by position)
df2.iloc[0:5, 0:3]           # Get rows 0-4, columns 0-2 (by position)
```

---

### Exam Tip ✅
> **Use `.loc` for labels, use `.iloc` for integer positions.**

## Understanding `df2.dropna()`

`df2.dropna()` means:

> **Remove any row that contains at least one missing value (NaN).**

### Simple Example

Suppose we have this DataFrame:

| A   | B   |
| --- | --- |
| 1   | 4   |
| NaN | 5   |
| 3   | NaN |

If we run: `df2.dropna()`

### What happens?

* Row 2 ❌ (because A is missing)
* Row 3 ❌ (because B is missing)
* Only Row 1 remains ✅

Result:

| A | B |
| - | - |
| 1 | 4 |

---

### Important Note

* `dropna()` works on **rows by default**
* If **even one NaN** exists in a row, that row is removed

---

### Alternative: Count missing values instead of removing


This counts how many missing values are in each column **without removing rows**.

## 17. Instead of removing data, you want to replace all missing values in the 'Target' column with the mean price

**Answer:** Use `df2['Target'] = df2['Target'].fillna(df2['Target'].mean())`

````python
# Load the data
df2 = pd.read_csv("../data/housing.csv")
````

## Method: Fill missing values in Target column with the mean

To replace missing values in the Target column with its average value:

````python
# Fill missing values in Target column with the mean of the column
# Step 1: Calculate the mean value
mean_target = df2["Target"].mean()

# Step 2: Replace all NaN with that mean
df2["Target"] = df2["Target"].fillna(mean_target)

# Verify no missing values remain
print(f"Missing values after filling: {df2['Target'].isna().sum()}")
````

````python
# Demonstrate the fillna() method with a simple example
# Create a Series with some NaN values
s = pd.Series([1, np.nan, 3, np.nan, 5])
print("Original Series:")
print(s)

# Fill NaN with the mean
print("\nAfter fillna(mean):")
print(s.fillna(s.mean()))
````

````python
# If MedHouseVal column exists, create Target column from it
if "MedHouseVal" in df2.columns and "Target" not in df2.columns:
    df2["Target"] = df2["MedHouseVal"]

# Fill any missing values in Target with the column mean
if "Target" in df2.columns:
    df2["Target"] = df2["Target"].fillna(df2["Target"].mean())
    print("Target column has been processed successfully")
````

## 29. Why is using df.loc or df.iloc preferred over chained indexing (e.g., df['col'][0])?

**Answer: D) All of the above**

Reasons:
- **Clarity** - More readable code
- **Performance** - Slightly faster
- **Avoids SettingWithCopyWarning** - Prevents confusing warnings when modifying data
- **Best Practice** - Recommended by pandas developers

**Recommended approaches:**
1. `.loc` for label-based indexing (by column/row names)
2. `.iloc` for position-based indexing (by integer positions)
3. Avoid chained indexing like `df['col'][0]`

````python
# Chained indexing (NOT recommended - can cause SettingWithCopyWarning)
value1 = df2['MedInc'][0]
print("Value using chained indexing:", value1)
````

````python
# Better approach 1: Use .iloc on the Series
value2 = df2['MedInc'].iloc[0]
print("Value using .iloc on Series:", value2)
````

````python
# Best approach 2: Use .loc (label-based indexing - RECOMMENDED)
value3 = df2.loc[0, 'MedInc']
print("Value using .loc (RECOMMENDED):", value3)
````

````python
# Best approach 3: Use .iloc (position-based indexing)
value4 = df2.iloc[0, 0]
print("Value using .iloc:", value4)

# Verify all methods return the same value
print("\nAll methods return the same values as shown above")
````
