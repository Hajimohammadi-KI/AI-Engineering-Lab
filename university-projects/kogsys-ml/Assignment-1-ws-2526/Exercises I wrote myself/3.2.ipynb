````python
import torch
import torch.nn as nn
print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA verfügbar: {torch.cuda.is_available()}")
````

````python
"""
Question 3.2: Calculate the number of parameters in a neural network

This network consists of the following layers:
1. Conv2d: 64 filters, 3x3 kernel, no bias
2. ReLU: no parameters
3. MaxPool2d: no parameters
4. Flatten: no parameters
5. Linear: 1024 input features, 256 output features, no bias

Formulas for calculating parameters:
- Conv2d: (kernel_size × kernel_size × in_channels × out_channels) + bias
- Linear: (in_features × out_features) + bias
"""

# Define the network according to the question
model = nn.Sequential(
    # Convolution layer: 64 input channels, 64 output channels, 3x3 kernel, no bias
    nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, stride=1, padding=1, bias=False),
    # ReLU activation function - no learnable parameters
    nn.ReLU(),
    # MaxPooling layer - no learnable parameters
    nn.MaxPool2d(kernel_size=2, stride=2),
    # Flatten - converts multi-dimensional tensor to 1D - no parameters
    nn.Flatten(),
    # Linear layer (Fully Connected): 1024 inputs, 256 outputs, no bias
    nn.Linear(in_features=1024, out_features=256, bias=False),
)

print("=" * 60)
print("Calculating Neural Network Parameters")
print("=" * 60)

# --- Calculate Conv2d parameters ---
# Formula: kernel_size × kernel_size × in_channels × out_channels
# In this question: 3 × 3 × 64 × 64 = 36,864
conv_params = 3 * 3 * 64 * 64  # kernel_size=3, in_channels=64, out_channels=64
print(f"\n🔹 Conv2d Layer:")
print(f"   Formula: kernel_size² × in_channels × out_channels")
print(f"   Calculation: 3 × 3 × 64 × 64 = {conv_params:,}")

# --- Calculate Linear layer parameters ---
# Formula: in_features × out_features
# In this question: 1024 × 256 = 262,144
linear_params = 1024 * 256  # in_features=1024, out_features=256
print(f"\n🔹 Linear Layer:")
print(f"   Formula: in_features × out_features")
print(f"   Calculation: 1024 × 256 = {linear_params:,}")

# --- Layers without parameters ---
print(f"\n🔹 Layers without parameters:")
print(f"   - ReLU: 0 parameters")
print(f"   - MaxPool2d: 0 parameters")
print(f"   - Flatten: 0 parameters")

# --- Total parameters ---
total_params = conv_params + linear_params
print(f"\n{'=' * 60}")
print(f"✅ Total parameters: {conv_params:,} + {linear_params:,} = {total_params:,}")
print(f"{'=' * 60}")

# --- Verification with PyTorch ---
print(f"\n📊 Verification with PyTorch:")
pytorch_total = sum(p.numel() for p in model.parameters())
print(f"   Number of parameters via model.parameters(): {pytorch_total:,}")

# Display details of each layer
print(f"\n📋 Parameter details for each layer:")
for name, param in model.named_parameters():
    print(f"   {name}: shape = {list(param.shape)}, count = {param.numel():,}")
````

# Neural Network Layer Reference Guide

## 1. Core Layers Overview

| Layer      | Definition                                                        | Formula                        | Parameters (with bias)         | Use Case                                      |
|------------|-------------------------------------------------------------------|--------------------------------|-------------------------------|------------------------------------------------|
| Conv2d     | 2D Convolutional layer that applies learned filters               | Y = X * W + b                  | (K x K x Cin x Cout) + Cout    | Feature extraction from images                 |
| ReLU       | Rectified Linear Unit - introduces non-linearity                  | f(x) = max(0, x)               | 0                             | Adds non-linearity, prevents vanishing gradient|
| MaxPool2d  | 2D Max Pooling - downsamples by max value in each window          | Y = max(X_region)              | 0                             | Reduces spatial dimensions, translation invariance|
| Flatten    | Reshapes multi-dimensional tensor into 1D vector                  | Y = reshape(X, -1)             | 0                             | Connects CNN layers to fully connected layers  |
| Linear     | Fully Connected layer - every input to every output               | Y = XW^T + b                   | (in x out) + out              | Classification, regression, final decision     |

---

## 2. Key Concepts

| Concept      | Definition                                   | Example/Formula                | Impact                                 |
|--------------|----------------------------------------------|-------------------------------|----------------------------------------|
| Stride       | Step size for sliding filter/kernel          | stride=1: move 1 pixel         | Larger stride → smaller output size    |
| Padding      | Zeros added around input borders             | padding=1: add 1 pixel border  | Preserves spatial dimensions           |
| Downsampling | Reducing spatial dimensions of feature maps  | Pooling or stride > 1          | Reduces computation, increases receptive field |
| Kernel/Filter| Small matrix of learnable weights            | 3x3, 5x5, 7x7                  | Determines feature detection pattern   |

---

## 3. Output Size Formulas

For Conv2d and Pooling:
Output_size = floor((Input_size - Kernel_size + 2 * Padding) / Stride) + 1

Examples:
| Input | Kernel | Stride | Padding | Output |
|-------|--------|--------|---------|--------|
| 32x32 | 3x3    | 1      | 1       | 32x32  |
| 32x32 | 3x3    | 1      | 0       | 30x30  |
| 32x32 | 2x2    | 2      | 0       | 16x16  |

---

## 4. Parameter Count Formulas

| Layer                | Without Bias           | With Bias                    |
|----------------------|-----------------------|------------------------------|
| Conv2d               | K^2 x Cin x Cout      | K^2 x Cin x Cout + Cout      |
| Linear               | in_features x out_features | in_features x out_features + out_features |
| ReLU, MaxPool, Flatten | 0                   | 0                            |

---

## 5. Comparison Table (Unterschiede)

| Feature                    | Conv2d | MaxPool2d | Linear |
|----------------------------|--------|-----------|--------|
| Learnable?                 | Ja     | Nein      | Ja     |
| Preserves Spatial Structure?| Ja     | Ja        | Nein   |
| Reduces Dimensions?        | Optional (stride) | Ja | Nein |
| Local vs Global            | Local  | Local     | Global |
| Main Purpose               | Feature extraction | Downsampling | Decision making |

---

## 6. Activation Functions Comparison

| Function   | Formula         | Range         | Pros                        | Cons                  |
|------------|----------------|--------------|-----------------------------|-----------------------|
| ReLU       | max(0, x)      | [0, ∞)       | Fast, no vanishing gradient | Dead neurons (x<0)    |
| Sigmoid    | 1/(1+exp(-x))  | (0, 1)       | Smooth, probabilistic       | Vanishing gradient    |
| Tanh       | (exp(x)-exp(-x))/(exp(x)+exp(-x)) | (-1, 1) | Zero-centered              | Vanishing gradient    |
| Leaky ReLU | max(0.01x, x)  | (-∞, ∞)      | No dead neurons             | Extra hyperparameter  |

---

    ↓            ↓          ↓          ↓            ↓           ↓
 (H,W,C)    Features   Non-linear  Downsample    1D Vector   Classes
            Extraction  Activation  (reduce size)  (flatten)  (predict)


# 🎯 درک عمیق پارامترهای Conv2d (توضیح فارسی)

## ❓ سوال اصلی: پارامتر در Conv2d چیست؟

پارامترها = **وزن‌های کرنل‌ها** (Kernel Weights)

> هر **فیلتر** یک کرنل 3×3 دارد **برای هر کانال ورودی**

---

## 📐 فرمول کلی

$$\text{Parameters} = \underbrace{\text{out\_channels}}_{\text{تعداد فیلترها}} \times \underbrace{\text{in\_channels}}_{\text{کانال‌های ورودی}} \times \underbrace{k_H \times k_W}_{\text{اندازه کرنل}}$$

---

## 🔢 مثال ما: in=64, out=64, kernel=3×3

### مرحله ۱: یک کرنل چند وزن دارد؟
```
┌───┬───┬───┐
│ w │ w │ w │     3 × 3 = 9 وزن
├───┼───┼───┤
│ w │ w │ w │
├───┼───┼───┤
│ w │ w │ w │
└───┴───┴───┘
```
**جواب: 9 وزن**

---

### مرحله ۲: یک فیلتر چند وزن دارد؟

هر فیلتر باید روی **همه‌ی 64 کانال ورودی** کار کند:

```
فیلتر شماره 1:
├── کرنل برای کانال ورودی 1:  [3×3] = 9 وزن
├── کرنل برای کانال ورودی 2:  [3×3] = 9 وزن
├── کرنل برای کانال ورودی 3:  [3×3] = 9 وزن
│   ...
└── کرنل برای کانال ورودی 64: [3×3] = 9 وزن
                                    ─────────
                              جمع: 64 × 9 = 576 وزن
```

**جواب: 576 وزن برای هر فیلتر**

---

### مرحله ۳: کل لایه چند وزن دارد؟

ما **64 فیلتر** داریم (out_channels=64):

```
فیلتر  1: 576 وزن
فیلتر  2: 576 وزن
فیلتر  3: 576 وزن
   ...
فیلتر 64: 576 وزن
─────────────────
جمع کل: 64 × 576 = 36,864 وزن
```

---

## 🖼️ تصویر کامل

```
                    ╔═══════════════════════════════════════╗
                    ║         64 فیلتر (out_channels)       ║
                    ╚═══════════════════════════════════════╝
                                      │
        ┌─────────────────────────────┼─────────────────────────────┐
        ▼                             ▼                             ▼
   ┌─────────┐                   ┌─────────┐                   ┌─────────┐
   │ فیلتر 1 │                   │ فیلتر 2 │       ...        │فیلتر 64 │
   └────┬────┘                   └────┬────┘                   └────┬────┘
        │                             │                             │
   ┌────┴────┐                   ┌────┴────┐                   ┌────┴────┐
   │64 کرنل │                   │64 کرنل │                   │64 کرنل │
   │ (3×3)  │                   │ (3×3)  │                   │ (3×3)  │
   └─────────┘                   └─────────┘                   └─────────┘
   = 64×9=576                    = 64×9=576                    = 64×9=576

                    ═══════════════════════════════════════
                    جمع کل = 64 × 576 = 36,864 پارامتر
```

---

## 📊 جدول محاسبه گام به گام

| مرحله | محاسبه | نتیجه |
|-------|--------|-------|
| وزن‌های یک کرنل | $3 \times 3$ | **9** |
| وزن‌های یک فیلتر | $64 \times 9$ | **576** |
| وزن‌های کل Conv2d | $64 \times 576$ | **36,864** |
| Bias (غیرفعال) | $0$ | **0** |
| **مجموع** | $36,864 + 0$ | **36,864** ✅ |

---

## 🎓 مثال ساده‌تر برای درک بهتر

فرض کن:
- in_channels = **2** (مثل تصویر سیاه‌سفید با 2 کانال)
- out_channels = **3** (می‌خوای 3 نوع feature استخراج کنی)  
- kernel = **3×3**

```
فیلتر 1: 2 کرنل × 9 وزن = 18 وزن
فیلتر 2: 2 کرنل × 9 وزن = 18 وزن
فیلتر 3: 2 کرنل × 9 وزن = 18 وزن
────────────────────────────────
جمع: 3 × 2 × 9 = 54 وزن
```

---

## 💡 قانون طلایی (حفظ کن!)

$$\boxed{\text{یک خروجی} = \text{یک فیلتر} = \text{(تعداد کانال ورودی)} \times \text{کرنل}}$$

یعنی:
- **هر کانال خروجی** ← یک فیلتر مجزا
- **هر فیلتر** ← برای هر کانال ورودی یک کرنل
- **هر کرنل** ← ماتریس وزن‌ها (مثلاً 3×3)

---

## ⚠️ نکته درباره Bias

اگر `bias=True` بود:
$$\text{Bias parameters} = \text{out\_channels} = 64$$

ولی در سوال ما `bias=False`، پس:
$$\text{Bias} = 0$$
