````python
from typing import Callable  # Import Callable for type hints
import torch.utils.data as tdata  # Import PyTorch Dataset utilities
import os  # For file and directory operations
from PIL import Image  # For image loading
import numpy as np  # For numerical operations (not strictly needed here)
from sklearn.model_selection import train_test_split  # For splitting train/val

class Fruits360(tdata.Dataset):
    '''
    torch dataset for the Fruits-360 dataset
    '''

    def __init__(self, root : str, split : str = 'train', transform : Callable = None, label_transform : Callable = None):
        ''' 9 Points
        Parameters
        ----------
        - root (str) : path to the to root directory of the dataset
        - split (str) : one of 'train', 'val', or 'test'. Should decide which split of the dataset to return.
        - transform (Callable) : transform callable to apply to the images
        - label_transform (Callable) : transform callable to apply to the labels
        '''
        # (1) Root-Verzeichnis als Attribut speichern

        self.root = root  # Store root directory
        self.split = split  # Store split type
        self.transform = transform  # Store image transform
        self.label_transform = label_transform  # Store label transform

        # (2) Bildpfade und Labels sammeln . Determine the correct folder for the split
        if split == 'test':
            data_dir = os.path.join(root, 'Test')  # Use Test folder for test split
        else:
            data_dir = os.path.join(root, 'Training')  # Use Training folder for train/val

        # (3) 10% Validierungssplit, reproduzierbar mit random_state .Collect all image paths and labels
        
        image_paths = []  # List to store image file paths
        labels = []  # List to store corresponding labels
        for class_name in sorted(os.listdir(data_dir)):  # Iterate over class folders
            # class_dir = os.path.join(data_dir, class_name)
            # if not os.path.isdir(class_dir):  # Skip if not a directory
            #     continue
            # for fname in os.listdir(class_dir):  # Iterate over files in class folder
            #     if fname.lower().endswith(('.jpg', '.jpeg', '.png')):  # Only image files
            #         image_paths.append(os.path.join(class_dir, fname))  # Add image path
            #         labels.append(class_name)  # Add label (class name)

            # corr:
            class_folder = os.path.join(data_dir, class_name)
            # Each class has subfolders with images
            for subfolder in os.listdir(class_folder):
                subfolder_path = os.path.join(class_folder, subfolder)
                if os.path.isdir(subfolder_path):
                    # Get all jpg files in this subfolder
                    for img_file in glob.glob(os.path.join(subfolder_path, '*.jpg')):
                        image_paths.append(img_file)
                        labels.append(class_name)


        # For train/val split, split data reproducibly
        if split in ['train', 'val']:
            # Use a fixed random state for reproducibility and stratify by label
            train_paths, val_paths, train_labels, val_labels = train_test_split(
                image_paths, labels, test_size=0.1, random_state=42, stratify=labels)
            if split == 'train':
                self.samples = list(zip(train_paths, train_labels))  # Store train samples
            else:
                self.samples = list(zip(val_paths, val_labels))  # Store val samples
        else:
            self.samples = list(zip(image_paths, labels))  # For test: use all samples

    def __getitem__(self, key : int) -> tuple:
        ''' 5 Points
        '''
        img_path, label = self.samples[key]  # Get image path and label for index
        image = Image.open(img_path).convert('RGB')  # Load image and ensure RGB
        if self.transform is not None:
            image = self.transform(image)  # Apply image transform if provided
        if self.label_transform is not None:
            label = self.label_transform(label)  # Apply label transform if provided
        return image, label  # Return tuple (image, label)

    def __len__(self) -> int:
        ''' 2 Points
        '''

        return len(self.samples)  # Return number of samples in split
````

### **(03.1.3)** Create `Transform` Classes and `Compose`

_For **6** Points_

**Implement the `Transform` callables that are passed to the dataset as the `transform` and `label_transform` parameters.**

1. **(2 Points)** For `label_transform`, create a class that converts the `str` labels from the dataframe to integer tensors. Depending on the implementation of the dataset, you can use the `codes` or `values` list containing the class names. The class must implement the `__call__` method, which defines the behavior of an object when it is called like a function.
2. **(4 Points)** For `transform`, the loaded `Image` should be converted to a tensor in PyTorch format `(c, h, w)`, and then scaled to the resolution `128x128`. Use the `transforms.Compose` container for this.

````python
import torch
from torchvision import transforms  # Import transforms for image preprocessing

class ConvertLabel(object):
    ''' 2 Points
    callable object converting a label to an integer
    '''
    def __init__(self, labels):
        # Store the list of class labels (e.g. ['Apple', 'Cherry', 'Tomato'])
        self.label_to_idx = {label: idx for idx, label in enumerate(labels)}  # Mapping label->int

    def __call__(self, sample : str) -> torch.Tensor:
        # Convert string label to integer index and return as tensor
        idx = self.label_to_idx[sample]
        return torch.tensor(idx, dtype=torch.long)

# Compose image transforms: resize to 128x128, convert to tensor (C,H,W), scale to [0,1]
fruits360_transforms = transforms.Compose([
    transforms.Resize((128, 128)),  # Resize image
    transforms.ToTensor()           # Convert to PyTorch tensor (C,H,W) and scale
])
````

### **(03.2.1)** Create Network

_For **10** Points_

**Implement a neural network consisting of blocks from the `torch.nn` module, the `BasicBlock` from `torchvision.models.resnet`, and possibly self-implemented or adapted blocks _(5 Points)_.**

**Further points for this task are awarded for complying with the following requirements:**

($<100,000$ -0 Points, $100,000 - 120,000$ -1 Point, $120,000-140,000$ -2 Points, $>140,000$ -3 Points).
- **(2 Points):** The model should achieve an accuracy of at least $.55$ on the test split of the dataset (within the framework of the requirements of the next task). Points will be deducted depending on the amount of shortfall
($>.55$ -0 Points, $.55-.5$ -1 Point, $<.5$ -2 Points).

````python
from torch import nn
from torch.nn import functional as F
from torchvision.models.resnet import BasicBlock

# Hey! You should really read the code of "BasicBlock" to get a
# better understanding of the architecture!
````

````python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        # Convolutional Block 1
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        # BasicBlock from ResNet (input: 16 channels, output: 16 channels)
        self.block1 = BasicBlock(16, 16)
        # Convolutional Block 2
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        # BasicBlock from ResNet (input: 32 channels, output: 32 channels)
        self.block2 = BasicBlock(32, 32)
        # Convolutional Block 3
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        # Global Average Pooling
        self.gap = nn.AdaptiveAvgPool2d((1, 1))
        # Fully Connected Layers
        self.fc1 = nn.Linear(64, 32)
        self.fc2 = nn.Linear(32, 3)  # 3 classes: Apple, Cherry, Tomato

    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.block1(x)
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.block2(x)
        x = F.relu(self.bn3(self.conv3(x)))
        x = self.gap(x)
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
````

### **(03.2.2)** Train Network

_For **23** Points_

**Implement a training loop for the network designed above. Points for this task are awarded for the implementation of the following components:**

1. **(11 Points) Training General.** Generally, a training loop for a `PyTorch` network requires an `Optimizer` and a loss function. During training...

   - the model should be set to training mode

3. **(3 Points) Learning-Rate Scheduling.** Using the `torch.optim.lr_scheduler.ReduceLROnPlateau` learning rate scheduler, the learning rate of the optimizer should be reduced when a metric no longer improves. Choose a meaningful metric and patience and build the learning rate scheduler into your training loop.

Choose a meaningful metric and patience and build a condition for early stopping into your training loop. ##### **Performance Note: [Google Colab](https://colab.research.google.com)**

***Training networks of this size is very computationally intensive. When training on the CPU, the training time is so long that experimenting with different architectures is made difficult (because one has to wait very long for results).
````python
torch.manual_seed(42)
import copy
import numpy as np
from tqdm import tqdm
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau

device = 'cuda' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu'

# Instantiate model and move to device
model = Model().to(device)

# Optimizer and loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

# Learning rate scheduler: Reduce LR if val_loss does not improve
# scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3) # corr

# Early stopping parameters
early_stopping_patience = 7
best_val_acc = 0.0
best_model_wts = copy.deepcopy(model.state_dict())
epochs_no_improve = 0
num_epochs = 25
train_losses = []
val_losses = []
val_accuracies = []

for epoch in range(num_epochs):
    print(f'\nEpoch {epoch+1}/{num_epochs}')
    model.train()
    running_loss = 0.0
    total = 0
    correct = 0
    for images, labels in tqdm(train_loader, desc='Training', leave=False):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    train_losses.append(epoch_loss)
    print(f'Train Loss: {epoch_loss:.4f} | Train Acc: {epoch_acc:.4f}')

    # Validation
    model.eval()
    val_running_loss = 0.0
    val_total = 0
    val_correct = 0
    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc='Validation', leave=False):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_running_loss += loss.item() * images.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)
    val_loss = val_running_loss / val_total
    val_acc = val_correct / val_total
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)
    print(f'Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}')

    # Step scheduler on validation loss
    scheduler.step(val_loss)

    # Early stopping and checkpointing (monitoring val_acc)
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        epochs_no_improve = 0
        print('Validation accuracy improved, saving model checkpoint.')
        torch.save(model.state_dict(), 'best_model.pt')
    else:
        epochs_no_improve += 1
        print(f'No improvement for {epochs_no_improve} epochs.')
        if epochs_no_improve >= early_stopping_patience:
            print('Early stopping triggered! Restoring best model weights.')
            model.load_state_dict(best_model_wts)
            break

# After training, ensure best model is loaded
model.load_state_dict(best_model_wts)
print(f'Best validation accuracy: {best_val_acc:.4f}')
````

### **(03.3.1)** How many input features must the first Linear Layer (Layer 14) of the network described above have? Also explain how the number is calculated.

_For **4** Points_

**Antwort:**

Die Anzahl der Eingabefeatures für die erste Linear-Schicht (Layer 14) entspricht der Anzahl der Ausgabewerte nach dem letzten Faltungsschritt und dem Flattening.

**Berechnungsschritte:**
1. **Eingabe:** 3 Kanäle, Bildgröße 96×96.
2. **Layer 1:** Conv2d(3, 24, kernel=5, stride=1, padding=2, groups=1, bias=False) → Ausgabe: 24×96×96
3. **Layer 2:** ReLU (keine Änderung)
4. **Layer 3:** MaxPool2d(kernel=2) → 24×48×48
5. **Layer 4:** Conv2d(24, 48, kernel=3, stride=1, padding=1, groups=3, bias=False) → 48×48×48
6. **Layer 5:** ReLU (keine Änderung)
7. **Layer 6:** Conv2d(48, 48, kernel=3, stride=2, padding=1, groups=1, bias=False) → 48×24×24
8. **Layer 7:** ReLU (keine Änderung)
9. **Layer 8:** Conv2d(48, 64, kernel=3, stride=1, padding=1, groups=4, bias=False) → 64×24×24
10. **Layer 9:** ReLU (keine Änderung)
11. **Layer 10:** MaxPool2d(kernel=2) → 64×12×12
12. **Layer 11:** Conv2d(64, 64, kernel=3, stride=1, padding=1, groups=8, bias=False) → 64×12×12
13. **Layer 12:** ReLU (keine Änderung)
14. **Layer 13:** Flatten → 64×12×12 = **9216** Features

**Antwort:** Die erste Linear-Schicht (Layer 14) benötigt **9216 Eingabefeatures**.

**Begründung:** Nach allen Convolution- und Pooling-Layern bleibt ein Tensor der Form (Batch, 64, 12, 12) übrig. Nach dem Flattening ergibt das 64×12×12 = 9216 Features pro Bild.

### **(03.3.3)** Calculate the number of parameters of the network described above. Also ensure that the calculation path is comprehensible.

_For 10 Points_

**Berechnung der Anzahl der Parameter:**

Wir berechnen die Parameter für jede Schicht einzeln. Es gibt **keine Bias-Terme**.

#### 1. Conv2d(3, 24, kernel=5, stride=1, padding=2, groups=1, bias=False)
- Filtergröße: $5 \times 5$
- Eingabekanäle: 3
- Ausgabekanäle: 24
- Gruppen: 1 (Standard)
- Parameter: $24 \times 3 \times 5 \times 5 = 1{,}800$

#### 2. Conv2d(24, 48, kernel=3, stride=1, padding=1, groups=3, bias=False)
- Filtergröße: $3 \times 3$
- Eingabekanäle pro Gruppe: $24/3 = 8$
- Ausgabekanäle pro Gruppe: $48/3 = 16$
- Jede Gruppe: $16 \times 8 \times 3 \times 3 = 1{,}152$
- Insgesamt: $1{,}152 \times 3 = 3{,}456$

#### 3. Conv2d(48, 48, kernel=3, stride=2, padding=1, groups=1, bias=False)
- Filtergröße: $3 \times 3$
- Eingabekanäle: 48
- Ausgabekanäle: 48
- Gruppen: 1
- Parameter: $48 \times 48 \times 3 \times 3 = 20{,}736$

#### 4. Conv2d(48, 64, kernel=3, stride=1, padding=1, groups=4, bias=False)
- Filtergröße: $3 \times 3$
- Eingabekanäle pro Gruppe: $48/4 = 12$
- Ausgabekanäle pro Gruppe: $64/4 = 16$
- Jede Gruppe: $16 \times 12 \times 3 \times 3 = 1{,}728$
- Insgesamt: $1{,}728 \times 4 = 6{,}912$

#### 5. Conv2d(64, 64, kernel=3, stride=1, padding=1, groups=8, bias=False)
- Filtergröße: $3 \times 3$
- Eingabekanäle pro Gruppe: $64/8 = 8$
- Ausgabekanäle pro Gruppe: $64/8 = 8$
- Jede Gruppe: $8 \times 8 \times 3 \times 3 = 576$
- Insgesamt: $576 \times 8 = 4{,}608$

#### 6. Linear(9216, 512, bias=False)
- Parameter: $9216 \times 512 = 4{,}718{,}592$

#### 7. Linear(512, 128, bias=False)
- Parameter: $512 \times 128 = 65{,}536$

#### 8. Linear(128, 32, bias=False)
- Parameter: $128 \times 32 = 4{,}096$

#### 9. Linear(32, 8, bias=False)
- Parameter: $32 \times 8 = 256$

---
**Gesamtsumme:**
- $1{,}800$ (Conv1)
- $3{,}456$ (Conv2)
- $20{,}736$ (Conv3)
- $6{,}912$ (Conv4)
- $4{,}608$ (Conv5)
- $4{,}718{,}592$ (FC1)
- $65{,}536$ (FC2)
- $4{,}096$ (FC3)
- $256$ (FC4)

**Summe:**
$$
1{,}800 + 3{,}456 + 20{,}736 + 6{,}912 + 4{,}608 + 4{,}718{,}592 + 65{,}536 + 4{,}096 + 256 = 4{,}825{,}992
$$
