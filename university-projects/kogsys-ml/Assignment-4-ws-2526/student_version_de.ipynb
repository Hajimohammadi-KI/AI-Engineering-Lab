### Q-Learning

Rudi kennt die optimalen Wege im Lager nicht im Voraus. Er muss durch wiederholtes Ausprobieren lernen, welche Entscheidungen langfristig zu einer effizienten Lieferung führen. Ziel ist es daher, mithile von Q-Learning für jede mögliche Aktion zu lernen, wie gut sie im Hinblick auf den langfistigen Gesamtreward ist.

Ziel dieser Aufgabe ist es, den Q-Learning Reinforcement-Learning Algorithmus zu implementieren. Der Algorithmus ist definiert wiefolgt:

- Initialisiere den Tabelleneintrag $\hat{Q}(s, a)$ mit $0$
- Betrachte den aktuellen Zustand $s$
- Wiederhole
  - Wähle eine Aktion $a$ und führe sie aus
  - Erhalte die direkte Belohnung $r$
  - Betrachte den neuen Zustand $s^\prime$
  - Aktualisiere den Tabelleneintrag für $\hat{Q}(s, a)$ wiefolgt: $$\hat{Q}(s, a) \leftarrow r + \gamma \max_{a^\prime}\hat{Q}(s^\prime,a^\prime)$$
  - $s \leftarrow s^\prime$

### **(04.3.4)** Viterbi-Algorithmus: Implementation

_Für **11** Punkte_

Implementiere den Viterbi-Algorithmus, um die wahrscheinlichste Zustandssequenz für eine gegebene Beobachtungssequenz zu finden.

Der Viterbi-Algorithmus ist definiert als:

**Initialisierung:**
$$\delta_1(i) = \pi_i \cdot b_i(o_1)$$
$$\psi_1(i) = 0$$

**Rekursion:**
$$\delta_t(j) = \max_{1 \leq i \leq N} [\delta_{t-1}(i) \cdot a_{ij}] \cdot b_j(o_t)$$
$$\psi_t(j) = \arg\max_{1 \leq i \leq N} [\delta_{t-1}(i) \cdot a_{ij}]$$

**Terminierung:**
$$X^* = \max_{1 \leq i \leq N} \delta_T(i)\eta_i$$
$$x_T^* = \arg\max_{1 \leq i \leq N} \delta_T(i)\eta_i$$

**Backtracking:**
$$x_t^* = \psi_{t+1}(x_{t+1}^*)$$

Implementiere die Klasse `ViterbiHMM` mit folgenden Methoden:

1. `__init__(self, states, observations, A, B, pi)`: Konstruktor, der die Parameter des HMM speichert
