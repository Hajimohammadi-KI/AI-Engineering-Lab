## **2 | Reinforcement Learning**

_For a total of **49** points_

LazyLogistics GmbH has a problem: Its employees no longer want to walk through the warehouse all day to fetch packages. So an autonomous transport robot was purchased. His name is Rudi.

Rudi works in a simplified warehouse modeled as a directed graph. The warehouse consists of aisles and nodes between which the robot can move stepwise. Every movement costs time and energy, and thus money. Management therefore looks very closely at his efficiency.

Rudi's mission is thus:

- He starts at a fixed position in the warehouse.
- He must first drive to a package and pick it up.
- Only after picking up the package may he drive to the shipping station and complete his task.

### Warehouse Layout

Even at the beginning of his task, RoboRudi faces a decision. From the starting point, several paths lead to the package. Part of the paths runs over flat ground. On these routes, Rudi can move quickly as they are easy to traverse. Movements on flat terrain are therefore comparatively cheap and are rewarded with a positive reward. Alternatively, Rudi can choose a path from the start that leads over stairs. On stairs, the robot moves slower and needs more energy. Accordingly, movements in this area are provided with low rewards.

Regardless of the chosen path, Rudi eventually reaches the warehouse area where the package is located. There he can pick up the package. Picking up the package costs time and is therefore associated with a negative reward.

After Rudi has picked up the package, he faces a decision again. From this point, two different paths lead to the shipping station. One path runs over stairs. This path is comparatively short, but every single movement is strenuous and associated with higher costs. The alternative path runs over flat ground. On this path, Rudi can move faster, but it is significantly longer and requires more individual steps.

At the end of this path, Rudi reaches the shipping station. There he successfully delivers the package. Successful delivery represents the actual goal of the task and is thus honored with a high positive reward. After delivery, the task is complete. The shipping station is a terminal state.

### Abstract Modeling

![Domain.png](images/Domain.png)

To solve Rudi's task algorithmically, the warehouse is modeled as a directed graph. Each node corresponds to a state of the robot, describing its position in the warehouse as well as implicitly the package state (without package or with package).

Edges between states correspond to possible actions of the robot, e.g., movements or picking up the package. Each action leads deterministically to exactly one subsequent state.

Each action is linked to a fixed reward (annotated on the edges in the graph above), which describes the costs or benefits of the action:

- Movements on flat ground provide a small positive reward
- Movements over stairs provide a low or negative reward
- Picking up the package provides a negative reward
- Successfully delivering the package provides a high positive reward

Stairs and speed are not modeled as separate states, but exclusively mapped via reward values.

### Q-Learning

Rudi does not know the optimal paths in the warehouse in advance. He has to learn through repeated trial and error which decisions lead to efficient delivery in the long run. The goal is therefore to use Q-Learning to learn for every possible action how good it is regarding the long-term total reward.

The goal of this task is to implement the Q-Learning Reinforcement Learning algorithm. The algorithm is defined as follows:

- Initialize the table entry $\hat{Q}(s, a)$ with $0$
- Observe the current state $s$
- Repeat
  - Choose an action $a$ and execute it
  - Receive the immediate reward $r$
  - Observe the new state $s^\prime$
  - Update the table entry for $\hat{Q}(s, a)$ as follows: $$\hat{Q}(s, a) \leftarrow r + \gamma \max_{a^\prime}\hat{Q}(s^\prime,a^\prime)$$
  - $s \leftarrow s^\prime$

````python
from dataclasses import dataclass, field
from typing import Dict, List, Tuple, Optional, Set
import random

@dataclass
class Domain:
    """
    Domain representation for Rudi the robot's warehouse navigation.
    
    The warehouse is modeled as a directed graph where:
    - Nodes = States (robot positions)
    - Edges = Actions (movements between positions)
    - Edge weights = Rewards (costs/benefits of actions)
    
    States:
    - S_START: Starting position
    - F_PRE_1, F_PRE_2: Flat path before package (positive rewards: +2)
    - T_PRE: Stairs path before package (negative reward: -1)
    - PACKAGE: Package pickup location (picking up costs time: -3)
    - F_POST_1, F_POST_2, F_POST_3: Flat path after package (longer but +2 each)
    - T_POST: Stairs path after package (shorter but -2)
    - GOAL: Shipping station (terminal state, +10 reward)
    """
    
    # Graph structure: state -> list of (action_name, next_state, reward)
    graph: Dict[str, List[Tuple[str, str, float]]] = field(default_factory=dict)
    
    # All states in the domain
    states: Set[str] = field(default_factory=set)
    
    # Terminal states (absorbing states)
    terminal_states: Set[str] = field(default_factory=set)
    
    # Start state
    start_state: str = "S_START"
    
    def __post_init__(self):
        """Initialize the warehouse domain based on the problem description."""
        
        # Build the graph based on the domain diagram
        # Format: state -> [(action_name, next_state, reward), ...]
        
        self.graph = {
            # From START: can go to flat path or stairs
            "S_START": [
                ("go_flat", "F_PRE_1", 2),      # Flat ground, positive reward
                ("go_stairs", "T_PRE", -1),     # Stairs, negative reward
            ],
            
            # Flat path before package
            "F_PRE_1": [
                ("move_forward", "F_PRE_2", 2), # Continue on flat ground
            ],
            "F_PRE_2": [
                ("go_to_package", "PACKAGE", 2), # Reach package area
            ],
            
            # Stairs path before package (shorter but harder)
            "T_PRE": [
                ("go_to_package", "PACKAGE", -1), # Reach package via stairs
            ],
            
            # Package location: must pick up package
            "PACKAGE": [
                ("pickup_package", "PACKAGE_DONE", -3), # Picking up costs time
            ],
            
            # After picking up, choose path to goal
            "PACKAGE_DONE": [
                ("go_flat_post", "F_POST_1", 2),   # Longer flat path
                ("go_stairs_post", "T_POST", -2),  # Shorter stairs path
            ],
            
            # Flat path after package (longer but easier)
            "F_POST_1": [
                ("move_forward", "F_POST_2", 2),
            ],
            "F_POST_2": [
                ("move_forward", "F_POST_3", 2),
            ],
            "F_POST_3": [
                ("deliver", "GOAL", 10),          # Reach goal
            ],
            
            # Stairs path after package (shorter but harder)
            "T_POST": [
                ("deliver", "GOAL", 10),          # Reach goal via stairs
            ],
            
            # Goal is terminal state
            "GOAL": [],  # No actions from terminal state
        }
        
        # Extract all states
        self.states = set(self.graph.keys())
        for state, actions in self.graph.items():
            for _, next_state, _ in actions:
                self.states.add(next_state)
        
        # Define terminal states
        self.terminal_states = {"GOAL"}
    
    def get_actions(self, state: str) -> List[Tuple[str, str, float]]:
        """
        Get all possible actions from a given state.
        
        Returns: List of (action_name, next_state, reward) tuples
        """
        return self.graph.get(state, [])
    
    def get_reward(self, state: str, action: str) -> float:
        """Get the reward for performing an action in a state."""
        for act_name, next_state, reward in self.graph.get(state, []):
            if act_name == action:
                return reward
        return 0.0
    
    def get_next_state(self, state: str, action: str) -> Optional[str]:
        """Get the resulting state after performing an action."""
        for act_name, next_state, reward in self.graph.get(state, []):
            if act_name == action:
                return next_state
        return None
    
    def is_terminal(self, state: str) -> bool:
        """Check if a state is a terminal (absorbing) state."""
        return state in self.terminal_states
    
    def get_all_state_actions(self) -> List[Tuple[str, str]]:
        """Get all (state, action) pairs for the Q-table."""
        pairs = []
        for state, actions in self.graph.items():
            for action_name, _, _ in actions:
                pairs.append((state, action_name))
        return pairs
    
    def get_random_start_state(self) -> str:
        """Get a random non-terminal state for exploration."""
        non_terminal = [s for s in self.states if not self.is_terminal(s) and self.get_actions(s)]
        return random.choice(non_terminal)


# Test the domain
domain = Domain()
print("=== Domain Structure ===")
print(f"All states: {domain.states}")
print(f"Terminal states: {domain.terminal_states}")
print(f"\nGraph structure:")
for state, actions in domain.graph.items():
    if actions:
        print(f"  {state}:")
        for action, next_state, reward in actions:
            print(f"    --({action}, r={reward})--> {next_state}")
````

````python
import pandas as pd
import numpy as np
from warnings import simplefilter

# Suppress performance warnings for DataFrame column additions
simplefilter(action='ignore', category=pd.errors.PerformanceWarning)


class QLearn:
    """
    Q-Learning implementation for the warehouse navigation problem.
    
    The Q-Learning algorithm learns optimal action values through trial and error.
    This implementation uses exploration-only (no exploitation) since the state
    space is small enough.
    """
    
    # =====================================================================
    # 04.2.2 Constructor (3 Points)
    # =====================================================================
    
    def __init__(self, domain: Domain, gamma: float = 0.9):
        """
        Initialize the Q-Learning agent.
        
        Parameters:
            domain: The Domain object representing the warehouse
            gamma: Discount factor for future rewards (default: 0.9)
        """
        self.domain = domain          # Store the domain
        self.gamma = gamma            # Discount factor
        self.epoch = 0                # Current training epoch (starts at 0)
        self.q_table = self.init_df() # Initialize the Q-table
    
    # =====================================================================
    # 04.2.3 Helper Methods (14 Points)
    # =====================================================================
    
    def init_df(self) -> pd.DataFrame:
        """
        Initialize the Q-Table as a DataFrame.
        
        Creates a table with all (state, action) pairs as rows
        and "Epoch 0" as the initial column with all values = 0.
        
        Returns:
            DataFrame with MultiIndex (state, action) and initial Q-values of 0
        """
        # Get all state-action pairs from the domain
        state_actions = self.domain.get_all_state_actions()
        
        # Create MultiIndex for (state, action) pairs
        index = pd.MultiIndex.from_tuples(state_actions, names=['State', 'Action'])
        
        # Create DataFrame with initial Q-values of 0
        df = pd.DataFrame(index=index, columns=['Epoch 0'])
        df['Epoch 0'] = 0.0
        
        return df
    
    def apply_random_action(self, state: str) -> Tuple[Tuple[str, str], str, float]:
        """
        Select and execute a random action from the given state.
        
        Parameters:
            state: Current state of the agent
            
        Returns:
            Tuple of:
                - (state, action): The state-action pair for Q-table lookup
                - next_state: The resulting state
                - reward: The immediate reward received
        """
        # Get all possible actions from this state
        actions = self.domain.get_actions(state)
        
        if not actions:
            return None  # No actions available (terminal state)
        
        # Randomly select an action
        action_name, next_state, reward = random.choice(actions)
        
        return (state, action_name), next_state, reward
    
    def get_reward(self, state: str, action: str) -> float:
        """
        Get the direct reward for a state-action pair.
        
        Parameters:
            state: Current state
            action: Action to perform
            
        Returns:
            The immediate reward for this action
        """
        return self.domain.get_reward(state, action)
    
    def update(self, state_action: Tuple[str, str], new_q_value: float):
        """
        Update the Q-table entry for a state-action pair.
        
        Creates a new column for the current epoch if needed,
        then updates the specific Q-value.
        
        Parameters:
            state_action: (state, action) tuple
            new_q_value: The new Q-value (r + gamma * max(Q(s', a')))
        """
        col_name = f'Epoch {self.epoch}'
        
        # Create new column if it doesn't exist
        if col_name not in self.q_table.columns:
            self.q_table[col_name] = None  # Initialize with None (missing values)
        
        # Update the Q-value for this state-action pair
        self.q_table.loc[state_action, col_name] = new_q_value
    
    def finalize(self):
        """
        Fill missing Q-values at the end of an epoch.
        
        For all state-action pairs not explored in this epoch,
        copy the Q-value from the previous epoch.
        """
        current_col = f'Epoch {self.epoch}'
        prev_col = f'Epoch {self.epoch - 1}'
        
        if current_col in self.q_table.columns and prev_col in self.q_table.columns:
            # Find rows with missing values in current epoch
            mask = self.q_table[current_col].isna()
            # Fill with values from previous epoch
            self.q_table.loc[mask, current_col] = self.q_table.loc[mask, prev_col]
    
    def is_goal_state(self, state: str) -> bool:
        """
        Check if the given state is a terminal (goal) state.
        
        A state is terminal if it has no outgoing actions.
        
        Parameters:
            state: State to check
            
        Returns:
            True if the state is a terminal state
        """
        return self.domain.is_terminal(state)
    
    def get_max_q_value(self, state: str) -> float:
        """
        Get the maximum Q-value for any action from the given state.
        
        Looks at the previous epoch's Q-values.
        
        Parameters:
            state: State to get max Q-value for
            
        Returns:
            Maximum Q-value, or 0 if terminal state
        """
        if self.is_goal_state(state):
            return 0.0
        
        # Get the previous epoch column
        prev_col = f'Epoch {self.epoch - 1}'
        if prev_col not in self.q_table.columns:
            return 0.0
        
        # Get all Q-values for actions from this state
        try:
            q_values = self.q_table.loc[state, prev_col]
            if isinstance(q_values, pd.Series):
                return q_values.max()
            return float(q_values)
        except KeyError:
            return 0.0
    
    # =====================================================================
    # 04.2.4 Training (17 Points)
    # =====================================================================
    
    def do_epoch(self, start_state: str = None, gamma: float = None) -> List[str]:
        """
        Execute one training epoch of Q-Learning.
        
        1. Increment epoch counter
        2. Select random start state if not provided
        3. Run inner loop until goal state is reached
        4. Finalize Q-values for unexplored actions
        
        Parameters:
            start_state: Starting state (random if None)
            gamma: Discount factor (uses self.gamma if None)
            
        Returns:
            List of states visited (the path taken)
        """
        # Step 1: Increment epoch counter
        self.epoch += 1
        
        # Use default gamma if not specified
        if gamma is None:
            gamma = self.gamma
        
        # Step 2: Select random start state if not provided
        if start_state is None:
            current_state = self.domain.get_random_start_state()
        else:
            current_state = start_state
        
        # Track the path for debugging/visualization
        path = [current_state]
        
        # Step 3: Inner loop - run until reaching goal state
        while not self.is_goal_state(current_state):
            # Select random action
            result = self.apply_random_action(current_state)
            
            if result is None:
                break  # No actions available
            
            state_action, next_state, reward = result
            
            # Calculate new Q-value: Q(s,a) = r + gamma * max(Q(s', a'))
            max_future_q = self.get_max_q_value(next_state)
            new_q_value = reward + gamma * max_future_q
            
            # Update Q-table
            self.update(state_action, new_q_value)
            
            # Move to next state
            current_state = next_state
            path.append(current_state)
        
        # Step 4: Finalize - copy unexplored Q-values from previous epoch
        self.finalize()
        
        return path
    
    def train(self, n_epochs: int = 100, start_state: str = None, 
              gamma: float = None) -> pd.DataFrame:
        """
        Train the Q-Learning agent for multiple epochs.
        
        Parameters:
            n_epochs: Number of training epochs
            start_state: Starting state for each epoch (random if None)
            gamma: Discount factor
            
        Returns:
            The final Q-table
        """
        for i in range(n_epochs):
            path = self.do_epoch(start_state=start_state, gamma=gamma)
            
            # Print progress every 10 epochs
            if (i + 1) % 10 == 0:
                print(f"Epoch {i + 1}/{n_epochs} completed. Path length: {len(path)}")
        
        return self.q_table
    
    # =====================================================================
    # 04.2.5 Application (8 Points)
    # =====================================================================
    
    def optimal(self, start_state: str) -> Tuple[List[str], float]:
        """
        Find the optimal path from a given start state to the goal.
        
        Uses the latest Q-values to select the best action at each step.
        
        Parameters:
            start_state: State to start from
            
        Returns:
            Tuple of (path as list of states, total reward)
        """
        path = [start_state]
        total_reward = 0.0
        current_state = start_state
        
        # Get the latest epoch column
        latest_col = f'Epoch {self.epoch}'
        
        while not self.is_goal_state(current_state):
            # Get all actions from current state
            actions = self.domain.get_actions(current_state)
            
            if not actions:
                break
            
            # Find the action with highest Q-value
            best_action = None
            best_q = float('-inf')
            best_next_state = None
            best_reward = 0
            
            for action_name, next_state, reward in actions:
                try:
                    q_value = self.q_table.loc[(current_state, action_name), latest_col]
                    if q_value > best_q:
                        best_q = q_value
                        best_action = action_name
                        best_next_state = next_state
                        best_reward = reward
                except KeyError:
                    continue
            
            if best_action is None:
                break
            
            # Move to best next state
            total_reward += best_reward
            current_state = best_next_state
            path.append(current_state)
        
        return path, total_reward
    
    def paths(self) -> Dict[str, Tuple[List[str], float]]:
        """
        Find optimal paths from all possible starting states.
        
        Returns:
            Dictionary mapping start states to (optimal_path, total_reward)
        """
        all_paths = {}
        
        # Get all non-terminal states with outgoing actions
        start_states = [s for s in self.domain.states 
                       if not self.is_goal_state(s) and self.domain.get_actions(s)]
        
        for state in sorted(start_states):
            path, reward = self.optimal(state)
            all_paths[state] = (path, reward)
            print(f"From {state}: {' -> '.join(path)} (Total reward: {reward:.2f})")
        
        return all_paths


# =========================================================================
# Test the Q-Learning Implementation
# =========================================================================

print("=" * 70)
print("Q-LEARNING TRAINING")
print("=" * 70)

# Create domain and Q-Learning agent
domain = Domain()
agent = QLearn(domain, gamma=0.9)

# Train the agent
print("\nTraining with 100 epochs...")
q_table = agent.train(n_epochs=100, start_state="S_START", gamma=0.9)

# Display final Q-table
print("\n" + "=" * 70)
print("FINAL Q-TABLE (Last 5 epochs)")
print("=" * 70)
# Show only the last 5 epochs for readability
last_cols = [col for col in q_table.columns if col.startswith('Epoch')][-5:]
print(q_table[last_cols].to_string())

# Find and display optimal paths
print("\n" + "=" * 70)
print("OPTIMAL PATHS FROM ALL STATES")
print("=" * 70)
all_paths = agent.paths()

# Highlight the path from the true start state
print("\n" + "=" * 70)
print("OPTIMAL PATH FROM S_START:")
print("=" * 70)
path, reward = agent.optimal("S_START")
print(f"Path: {' -> '.join(path)}")
print(f"Total accumulated reward: {reward:.2f}")
````

### **(04.3.4)** Viterbi Algorithm: Implementation

_For **11** points_

Implement the Viterbi algorithm to find the most likely state sequence for a given observation sequence.

The Viterbi algorithm is defined as:

**Initialization:**
$$\delta_1(i) = \pi_i \cdot b_i(o_1)$$
$$\psi_1(i) = 0$$

**Recursion:**
$$\delta_t(j) = \max_{1 \leq i \leq N} [\delta_{t-1}(i) \cdot a_{ij}] \cdot b_j(o_t)$$
$$\psi_t(j) = \arg\max_{1 \leq i \leq N} [\delta_{t-1}(i) \cdot a_{ij}]$$

**Termination:**
$$X^* = \max_{1 \leq i \leq N} \delta_T(i)\eta_i$$
$$x_T^* = \arg\max_{1 \leq i \leq N} \delta_T(i)\eta_i$$

**Backtracking:**
$$x_t^* = \psi_{t+1}(x_{t+1}^*)$$

Implement the `ViterbiHMM` class with the following methods:

1. `__init__(self, states, observations, A, B, pi)`: Constructor that stores the HMM parameters
2. `viterbi(self, obs_seq)`: Execute the Viterbi algorithm and return the most likely state sequence and its probability

Test your implementation with the observation sequence `['FLAT', 'FLAT', 'PACKAGE', 'GOAL']`.

````python
import numpy as np
from typing import List, Tuple


class ViterbiHMM:
    """
    Hidden Markov Model with Viterbi algorithm for finding the most likely state sequence.
    
    The Viterbi algorithm uses dynamic programming to efficiently find the most
    probable sequence of hidden states given a sequence of observations.
    
    Key concepts:
    - delta[t][j]: Maximum probability of being in state j at time t
    - psi[t][j]: The state at time t-1 that led to delta[t][j]
    """

    def __init__(self, states: List[str], observations: List[str], 
                 A: np.ndarray, B: np.ndarray, pi: np.ndarray, eta: np.ndarray = None):  # 1 Point
        """
        Initialize the HMM.

        Parameters:
            states (list): List of state names, e.g., ['START', 'FLAT_PATH', ...]
            observations (list): List of possible observation symbols, e.g., ['FLAT', 'STAIRS', ...]
            A (np.array): Transition matrix (N x N) where A[i,j] = P(state_j | state_i)
            B (np.array): Emission matrix (N x M) where B[i,k] = P(obs_k | state_i)
            pi (np.array): Initial state distribution (N,)
            eta (np.array): Exit probabilities (N,) - probability of ending in each state
        """
        self.states = states                    # List of state names
        self.observations = observations        # List of observation symbols
        self.N = len(states)                    # Number of states
        self.M = len(observations)              # Number of observation types
        self.A = np.array(A, dtype=float)       # Transition matrix
        self.B = np.array(B, dtype=float)       # Emission matrix
        self.pi = np.array(pi, dtype=float)     # Initial state distribution
        
        # Exit probabilities (default to 1 for all states if not provided)
        if eta is None:
            self.eta = np.ones(self.N)
        else:
            self.eta = np.array(eta, dtype=float)
        
        # Create mappings for quick lookups
        self.state_to_idx = {s: i for i, s in enumerate(states)}
        self.obs_to_idx = {o: i for i, o in enumerate(observations)}

    def viterbi(self, obs_seq: List[str]) -> Tuple[List[str], float]:
        """
        Find the most likely state sequence for the given observation sequence.
        
        Uses the Viterbi algorithm:
        1. Initialization: delta_1(i) = pi_i * b_i(o_1)
        2. Recursion: delta_t(j) = max_i[delta_{t-1}(i) * a_ij] * b_j(o_t)
        3. Termination: Find the best final state
        4. Backtracking: Trace back through psi to find the path

        Parameters
        ----------
        obs_seq: list
            Sequence of observations, e.g., ['FLAT', 'FLAT', 'PACKAGE', 'GOAL']

        Returns
        -------
        tuple[list, float]
            [Most likely sequence of states, Probability of this sequence]
        """
        T = len(obs_seq)  # Length of observation sequence
        
        # Convert observations to indices
        obs_indices = [self.obs_to_idx[o] for o in obs_seq]
        
        # Initialize delta and psi matrices
        # delta[t, j] = max probability of reaching state j at time t
        # psi[t, j] = the previous state that maximized delta[t, j]
        delta = np.zeros((T, self.N))
        psi = np.zeros((T, self.N), dtype=int)
        
        # =================================================================
        # Step 1: Initialization (t = 0)
        # delta_1(i) = pi_i * b_i(o_1)
        # psi_1(i) = 0 (no previous state)
        # =================================================================
        first_obs = obs_indices[0]
        for i in range(self.N):
            delta[0, i] = self.pi[i] * self.B[i, first_obs]
            psi[0, i] = 0  # No predecessor at t=0
        
        # Print initialization step for debugging
        print(f"\n=== Viterbi Algorithm ===")
        print(f"Observation sequence: {obs_seq}")
        print(f"\nStep 1 (Initialization) - After observing '{obs_seq[0]}':")
        for i, state in enumerate(self.states):
            print(f"  delta[0][{state}] = pi[{state}] * B[{state}][{obs_seq[0]}] = {self.pi[i]:.4f} * {self.B[i, first_obs]:.4f} = {delta[0, i]:.6f}")
        
        # =================================================================
        # Step 2: Recursion (t = 1 to T-1)
        # delta_t(j) = max_i[delta_{t-1}(i) * a_ij] * b_j(o_t)
        # psi_t(j) = argmax_i[delta_{t-1}(i) * a_ij]
        # =================================================================
        for t in range(1, T):
            obs_idx = obs_indices[t]
            print(f"\nStep {t+1} (Recursion) - After observing '{obs_seq[t]}':")
            
            for j in range(self.N):
                # Find the maximum over all previous states
                max_val = -1
                max_state = 0
                
                for i in range(self.N):
                    val = delta[t-1, i] * self.A[i, j]
                    if val > max_val:
                        max_val = val
                        max_state = i
                
                # delta[t][j] = max_val * b_j(o_t)
                delta[t, j] = max_val * self.B[j, obs_idx]
                psi[t, j] = max_state
                
                if delta[t, j] > 0:
                    print(f"  delta[{t}][{self.states[j]}] = max_i(delta[{t-1}][i] * A[i][{self.states[j]}]) * B[{self.states[j]}][{obs_seq[t]}]")
                    print(f"    = {max_val:.6f} * {self.B[j, obs_idx]:.4f} = {delta[t, j]:.6f}  (from state {self.states[max_state]})")
        
        # =================================================================
        # Step 3: Termination
        # X* = max_i(delta_T(i) * eta_i)
        # x_T* = argmax_i(delta_T(i) * eta_i)
        # =================================================================
        # Apply exit probabilities
        final_probs = delta[T-1] * self.eta
        best_final_state = np.argmax(final_probs)
        best_prob = final_probs[best_final_state]
        
        print(f"\nStep {T+1} (Termination):")
        print(f"  Best final state: {self.states[best_final_state]} with probability {best_prob:.6f}")
        
        # =================================================================
        # Step 4: Backtracking
        # x_t* = psi_{t+1}(x_{t+1}*)
        # =================================================================
        # Reconstruct the best path
        best_path_indices = [0] * T
        best_path_indices[T-1] = best_final_state
        
        for t in range(T-2, -1, -1):
            best_path_indices[t] = psi[t+1, best_path_indices[t+1]]
        
        # Convert indices to state names
        best_path = [self.states[i] for i in best_path_indices]
        
        print(f"\nBacktracking:")
        print(f"  Most likely state sequence: {' -> '.join(best_path)}")
        print(f"  Sequence probability: {best_prob:.6f}")
        
        return best_path, best_prob


# =========================================================================
# Test the Viterbi Implementation
# =========================================================================

print("=" * 70)
print("VITERBI ALGORITHM TEST")
print("=" * 70)

# Define the HMM parameters
states = ['START', 'FLAT_PATH', 'STAIR_PATH', 'PACKAGE', 'GOAL']
observations = ['FLAT', 'STAIRS', 'PACKAGE', 'GOAL']

# Transition matrix A (from our solution)
A = np.array([
    [0, 0.5, 0.5, 0, 0],     # START
    [0, 0, 0, 1, 0],          # FLAT_PATH
    [0, 0, 0, 1, 0],          # STAIR_PATH
    [0, 0, 0, 0, 1],          # PACKAGE
    [0, 0, 0, 0, 1],          # GOAL
])

# Emission matrix B (from our solution)
p_correct = 0.8
p_wrong = 0.2 / 3  # ≈ 0.0667

B = np.array([
    [p_correct, p_wrong, p_wrong, p_wrong],    # START (FLAT terrain)
    [p_correct, p_wrong, p_wrong, p_wrong],    # FLAT_PATH (FLAT terrain)
    [p_wrong, p_correct, p_wrong, p_wrong],    # STAIR_PATH (STAIRS terrain)
    [p_wrong, p_wrong, p_correct, p_wrong],    # PACKAGE (PACKAGE terrain)
    [p_wrong, p_wrong, p_wrong, p_correct],    # GOAL (GOAL terrain)
])

# Initial state distribution (start at START)
pi = np.array([1, 0, 0, 0, 0])

# Exit probabilities (all equal for simplicity)
eta = np.array([1, 1, 1, 1, 1])

# Create the HMM
hmm = ViterbiHMM(states, observations, A, B, pi, eta)

# Test with the given observation sequence
print("\n" + "=" * 70)
print("Test Case: Observation sequence ['FLAT', 'FLAT', 'PACKAGE', 'GOAL']")
print("=" * 70)

obs_sequence = ['FLAT', 'FLAT', 'PACKAGE', 'GOAL']
best_path, probability = hmm.viterbi(obs_sequence)

print("\n" + "=" * 70)
print("FINAL RESULTS")
print("=" * 70)
print(f"Observation sequence: {obs_sequence}")
print(f"Most likely state sequence: {best_path}")
print(f"Probability: {probability:.8f}")

# Interpret the result
print("\n" + "=" * 70)
print("INTERPRETATION")
print("=" * 70)
print("""
The Viterbi algorithm found the most likely sequence of hidden states
that would produce the given observations:

1. FLAT observation → Most likely in START (where we begin, terrain is FLAT)
2. FLAT observation → Most likely in FLAT_PATH (moved to flat path, terrain is FLAT)
3. PACKAGE observation → Most likely in PACKAGE (arrived at package, terrain is PACKAGE)
4. GOAL observation → Most likely in GOAL (delivered package, terrain is GOAL)

This matches the expected path through the warehouse!
""")
````
